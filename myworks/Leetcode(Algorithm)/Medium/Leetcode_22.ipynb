Big takeaways from this question:
    
    A. I could utilize multiple 'if' statements seperately. (without if elif elif else relationship)
    B. By doing multiple 'if' statements, I could set more than one recursion. 
        i.e, from the output that is generated by a round of recursion,
             two different recursive conditions could be satisfied at the same time
             
        ex) when left = 1, right = 2; which satisfies two conditions i) if left and ii) right > left




```python
def generateParenthesis(n):
#     n = int and return List[str] 
    def generate(p, left, right, parens=[]):
       
        
        if left:
            generate(p + '(', left-1, right)
            
        if right > left:
            generate(p + ')', left, right-1)
            
        if not right:
            parens += p,
        
        
            
        return parens
    
    return generate('', n, n)
        
    
    
```
